```python
def name_out_file(input_fastq):

    input_fastq = list(input_fastq)
    name_output_file = "".join(input_fastq[:-6])
    
    return name_output_file
```

```python
def fix_set(gc_bounds, length_bounds):
    if isinstance(gc_bounds, int) == True:
        min_gc_bounds, max_gc_bounds = 0, gc_bounds
    else:
         min_gc_bounds, max_gc_bounds = gc_bounds
            
    if isinstance(length_bounds, int) == True:
        min_length_bounds, max_length_bounds = 0, length_bounds
    else:
        min_length_bounds, max_length_bounds = length_bounds
    
    return min_gc_bounds, max_gc_bounds, min_length_bounds, max_length_bounds
```

```python
def filtr_GC(line_1):
    status = False
    
    CG = (line_1.count("C") + line_1.count("G"))/(len((line_1))-1)*100
    
    if min_gc_bounds <= CG <= max_gc_bounds:
        status = True
    
    return status
```

```python
def filtr_len(line_1):
    status = False
    
    if min_length_bounds <= (len((line_1))-1) <= max_length_bounds:
        status = True
        
    return status
```

```python
def filtr_qual(line_3):
    ASCII_table = {
        "33":"0", "34":"1", "35":"2", "36":"3", "37":"4", "38":"5", "39":"6", "40":"7", "41":"8", "42":"9", 
        "43":"10", "44":"11", "45":"12", "46":"13", "47":"14", "48":"15", "49":"16", "50":"17", "51":"18", "52":"19", 
        "53":"20", "54":"21", "55":"22", "56":"23", "57":"24", "58":"25", "59":"26", "60":"27", "61":"28", "62":"29",
        "63":"30", "64":"31", "65":"32", "66":"33", "67":"34", "68":"35", "69":"36", "70":"37", "71":"38", "72":"39", "73":"40"        
    }
    
    status = False
    line_Q = []
    line_ASCII = []
    sum_Q_score = 0
    aver_Q_score = 0

    for sym in line_3:
        ASCII_code = ord(sym)
        line_ASCII.append(ASCII_code)                    
                
        line_ASCII = line_ASCII[:-1]
    for ASCII_code in line_ASCII:
        Q_score = ASCII_table[str(ASCII_code)]
        sum_Q_score += int(Q_score)
        aver_Q_score = sum_Q_score/len(line_ASCII)

    if aver_Q_score >= quality_threshold:
        status = True
    return status   
```

### Самостоятельные функции

```python
def filtr_GC(input_fastq, output_file_prefix, i=0, j=4, gc_bounds = (0, 100)):
    """
    Filter reads by GC composition
    :param: 
    :return: none
    """
    if isinstance(gc_bounds, int) == True:
        min_gc_bounds, max_gc_bounds = 0, gc_bounds
    else:
         min_gc_bounds, max_gc_bounds = gc_bounds
    
    with open(input_fastq) as file_in:
        num_lines_file_in = len(file_in.readlines())
        with open(output_file_prefix, "a") as file_out:
            with open(input_fastq, "r") as file_in:
                
                lines = file_in.readlines()[i:j]
                line_1 = lines[1]
                CG = (line_1.count("C") + line_1.count("G"))/(len((line_1))-1)*100

                if min_gc_bounds <= CG <= max_gc_bounds:
                    
                    for line in lines:
                        file_out.write(line)
                        
    if j <= (num_lines_file_in - 1):
        filtr_GC(input_fastq, output_file_prefix, i+4, j+4, gc_bounds)
    return
```

```python
def filtr_len(input_fastq, output_file_prefix, i=0, j=4, length_bounds=(0, 2**32)):
    """
    Filter reeds by length.
    :param: 
    :return:
    """
    if isinstance(length_bounds, int) == True:
        min_length_bounds, max_length_bounds = 0, length_bounds
    else:
        min_length_bounds, max_length_bounds = length_bounds
    
        
    with open(input_fastq) as file_in:
        num_lines_file_in = len(file_in.readlines())
        with open(output_file_prefix, "a") as file_out:
            with open(input_fastq, "r") as file_in:
                
                lines = file_in.readlines()[i:j]
                line_1 = lines[1]
                lenght = len((line_1))-1

                if min_length_bounds <= lenght <= max_length_bounds:
                    for line in lines:
                        file_out.write(line)
                        
    if j <= (num_lines_file_in - 1):
        filtr_len(input_fastq, output_file_prefix, i+4, j+4, length_bounds)
    return
    
```
